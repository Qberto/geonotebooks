```python
import pandas as pd
import pandas as pd
import numpy as np
from tqdm import tqdm
from glob import glob

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

pd.options.mode.chained_assignment = None 
```

```python
crashes = pd.read_pickle('crashes_weatherstn.pkl')
```

```python
year = 2010
```

```python
positive = crashes[(crashes.year == year)]
negative = pd.read_pickle('negative_{}.pkl'.format(year))
positive['class'] = 1
negative['class'] = 0
positive.shape, negative.shape
```

```python
samples = pd.concat([positive, negative], ignore_index=True)
samples.shape
```

```python
samples.head()
```

```python
samples.day = samples.day.astype(int)
samples.dayofweek = samples.dayofweek.astype(int)
samples.hour = samples.hour.astype(int)
samples.month = samples.month.astype(int)
samples.year = samples.year.astype(int)
samples.dayofyear = samples.dayofyear.astype(int)
```

```python
samples['TOT_C'] = samples['TOT_C'].astype(int)
samples['REGION'] = samples['REGION'].astype(int)
samples['SAF_INDEX'] = samples['SAF_INDEX'].astype(int)
samples['TOT_SEV_C'] = samples['TOT_SEV_C'].astype(int)
samples['CrashSeverity'] = samples['CrashSeverity'].astype(int)
samples['NUMBERVEHICLESINVOLVED'] = samples['NUMBERVEHICLESINVOLVED'].astype(int)
```

```python
samples = samples.sample(frac=1).reset_index(drop=True)
```

```python
samples.head()
```

```python
weather = pd.read_pickle('hourly_weather_{}.pkl'.format(year))
```

```python
weather['datetime'] = pd.to_datetime(weather.date_time, infer_datetime_format=True)
```

```python
weather['dayofyear'] = weather.datetime.dt.dayofyear
weather['hour'] = weather.datetime.dt.hour
```

```python
weather['key'] = weather['year'].map(str) + \
                 weather['dayofyear'].map(str) + \
                 weather['hour'].map(str) + weather['Station ID']
```

```python
samples['key'] = samples['year'].map(str) + \
                      samples['dayofyear'].map(str) + \
                      samples['hour'].map(str) + samples['Station ID']
```

```python
samples.shape
```

```python
len(weather.key.unique())
```

```python
len(weather)
```

```python
unique_weather = weather.drop_duplicates(subset='key')
```

```python
samples_weather = pd.merge(samples, unique_weather, on='key', how='left')
```

```python
samples_weather.shape
```

```python
samples_weather.head()
```

```python
samples_weather.to_pickle('samples_{}.pkl'.format(year))
```
