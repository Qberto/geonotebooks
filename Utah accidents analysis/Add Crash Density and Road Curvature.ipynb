```python
import math
import pandas as pd
from arcgis.features import SpatialDataFrame

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_pickle('crashes_roads_segments.pkl')
```

```python
sdf = SpatialDataFrame(data=df)
```

```python
def add_incident_sum(grp):
    count = grp['OBJECTID'].count()
    grp['num_crashes'] = count
    
    return grp

sdf = sdf.groupby('OBJECTID').apply(add_incident_sum)
```

```python
sdf['crash_density'] = sdf['num_crashes'] * 1000.0 / sdf.geometry.length
```

```python
sdf.head()
```

```python
sdf['length'] = sdf.geometry.length
```

```python
sdf['distance'] = sdf.geometry.apply(lambda x: math.sqrt(((x.first_point.get('x') - x.last_point.get('x'))**2) +
                                                         ((x.first_point.get('y') - x.last_point.get('y'))**2)))
```

```python
sdf['curvature'] = sdf['length'] / sdf['distance']
```

```python
sdf.head()
```

```python
sdf.to_pickle('crashes_density_curvature.pkl')
```

```python
density_curvature = sdf[['OBJECTID', 'num_crashes', 'crash_density', 'length','curvature']]
```
