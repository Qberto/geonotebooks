# Utah Road Safety Index
Safety Index Score

Service Description: The Safety Index offers a statewide comparison of UDOT roadways, taking into account the different traffic patterns and volumes experienced in urban and rural areas. The Safety Index is a combination of four, equally weighted safety analysis sub-scores: Crash Rate Score, Severe Crash Rate Score, Crashes per Mile Score, Severe Crashes per Mile Score. The Safety Index is reported on a 0 to 10 scale, with 10 representing the worst conditions. The data reflect crashes from 2011 through 2013. For questions on this data please contact Scott Jones at wsjones@utah.gov. Data in this map is protected under 23 USC 409.
